## Fix currency so that we have integers and not strings

````python
def fix_currency(x):
    if isinstance(x, str):
        return(x.replace('$',"").replace(",",''))
    return x
````

````python
years = ["pci"+str(year) for year in range(1959,2009, 10)]
````

````python
years
````

````python
for year in years:
    df[year] = df[year].apply(fix_currency).astype('int')
````
